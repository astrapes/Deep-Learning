```python
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
seed=2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 준비된 수술 환자 데이터를 불러들임
data_set=np.loadtxt("../dataset/ThoraricSurgery.csv",delimiter=",")
data_set[:2,:]
```

```python
# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
X=data_set[:,0:17]
Y=data_set[:,17]
X
```

```python
# 딥러닝 모델 설정
# Input 17개, Hidden layer 30개, Output 0 or 1
model=Sequential([
    Dense(30,input_shape=(17,), activation="relu"),
    Dense(1, activation="sigmoid")
])
model.summary()
```

```python
# 딥러닝 환결 설정 (오차함수, 최적화 함수)
model.compile(loss="binary_crossentropy",optimizer="adam",
              metrics=["accuracy"])
```

```python
# 딥러닝 실행
model.fit(X,Y,epochs=20,batch_size=10)
```

```python
# 결과 출력
# evaluate() returns the loss value & metrics values for the model
print("\n Accuracy : %.4f"%(model.evaluate(X,Y,verbose=2))[1])
```

## Model 2

```python
model2=Sequential()
model2.add(Dense(24,input_shape=(17,), activation="relu"))
model2.add(Dense(8, activation="relu"))
model2.add(Dense(1, activation="sigmoid"))
model2.summary()
```

```python
# 딥러닝 환경 설정 (오차함수, 최적화함수)
model2.compile(loss="binary_crossentropy",optimizer="adam",
               metrics=["accuracy"])
```

```python
model2.fit(X, Y, epochs=500, batch_size=30)
```

```python
# 결과 출력
# evaluate() returns the loss value & metrics values for the model 
print("\n Accuracy: %.4f" % (model2.evaluate(X, Y, verbose=2))[1])
```
